Axes labels,legend,grid

```python
import matplotlib.pyplot as plt
%matplotlib inline

days = [1,2,3,4,5,6,7]
max = [50,51,52,48,47,49,46]
min = [43,42,40,44,33,35,37]
avg = [45,48,48,46,40,42,41]

plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max,label="Max")
plt.plot(days,min,label="Min")
plt.plot(days,avg,label="Avg")

plt.legend(loc="best")

plt.grid()
```

Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

c = np.array(['Google','Amazon','Apple','Tesla'])
r = np.array([98,136,89,65])
p = np.array([40,4,34,12])

plt.ylabel("revenue")
plt.title("US Tech")

plt.bar(c,r,width=0.4,label="Revenue")
plt.bar(c,p,width=0.4,label="Profit")
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

b_s_m = [113,85,90,150,149,88,93,115,135,80,77,82,129]
b_s_w = [67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.xlabel("Sugar Range")
plt.ylabel("Total no of Patients")
plt.title("Blood Sugar Analysis")

plt.hist([b_s_m , b_s_w], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'], label=['men','women'])
plt.legend()
```
