Linestyle

```python
# The keyword argument 'linestyle', or shorter 'ls'
# Use a dotted line
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,7,5,10])
plt.plot(y, ls = 'dotted') # also use ':'
plt.show()
```

```python
# Use a dashed line
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,7,5,10])
plt.plot(y, ls = 'dashed') #also use '--'
plt.show()
```

Line Color

```python
# the keyword argument 'color' or the shorter 'c'
# Set the line color to green
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,7,5,10])
plt.plot(y,c='g')
plt.show()
```

```python
# Use Hexadecimal color values
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,7,5,10])
plt.plot(y,c='#4CAF50')
plt.show()
```

Line Width

```python
# The keyword argument 'linewidth' or the shorter 'lw'
# Plot with a 15.5pt wide line
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,7,5,10])
plt.plot(y,lw='15.5')
plt.show()
```

Multiple Lines

```python
# Draw two lines by specifying a plt.plot() function for each line
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,9,4,8])
y = np.array([2,7,5,10])
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
# Draw two lines by specifiyng the x- and y-point values for both lines
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```
