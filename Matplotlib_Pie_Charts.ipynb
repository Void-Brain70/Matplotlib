Creating Pie Charts

```python
# the pie() function to draw pie charts
# A simple pie chart
import matplotlib.pyplot as plt
import numpy as np

p = np.array([45,25,10,20])

plt.pie(p)
plt.show()
```

```python
# The pie chart with the 'label' parameter
import matplotlib.pyplot as plt
import numpy as np

p = np.array([45,25,10,20])
l =["CSE","EEE","CE","ARCH"]

plt.pie(p,labels=l)
plt.show()
```

```python
# The start angle by specifying a 'startangle' parameter
import matplotlib.pyplot as plt
import numpy as np

p = np.array([45,25,10,20])
l =["CSE","EEE","CE","ARCH"]

plt.pie(p,labels=l,startangle=180)
plt.show()
```

```python
# The 'explode' parameter 
import matplotlib.pyplot as plt
import numpy as np

p = np.array([45,25,10,20])
l =["CSE","EEE","CE","ARCH"]
e = [0.3,0,0.1,0]

plt.pie(p,labels=l,explode=e)
plt.show()
```

```python
# The 'shadows' parameter is 'True'
import matplotlib.pyplot as plt
import numpy as np

p = np.array([45,25,10,20])
l =["CSE","EEE","CE","ARCH"]
e = [0.3,0,0.1,0]

plt.pie(p,labels=l,explode=e,shadow=True)

plt.show()
```

```python
# The colors parameter
# Specify a new color for each wedge
import matplotlib.pyplot as plt
import numpy as np

p = np.array([45,25,10,20])
l =["CSE","EEE","CE","ARCH"]
c = ['g','m','y','w']

plt.pie(p,labels=l,colors=c)

plt.show()
```

```python
# The legend() function
import matplotlib.pyplot as plt
import numpy as np

p = np.array([45,25,10,20])
l =["CSE","EEE","CE","ARCH"]

plt.pie(p,labels=l)

plt.legend()

plt.show()
```

```python
# The title parameter to the legend() function
import matplotlib.pyplot as plt
import numpy as np

p = np.array([45,25,10,20])
l =["CSE","EEE","CE","ARCH"]

plt.pie(p,labels=l)
plt.legend(title="Department")
plt.show()
```
