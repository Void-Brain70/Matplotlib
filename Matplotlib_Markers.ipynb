Markers

```python
# Mark each point with a circle
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,10,8])
plt.plot(y,marker='o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,10,8])
plt.plot(y,marker='*')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,10,8])
plt.plot(y,marker='+')
plt.show()
```

Format Strings fmt

```python
# marker|line|color
# Mark each point with a circle
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,10,8])
plt.plot(y,'*:b')
plt.show()
```

Line Reference('-'Solid,':'Dotted,'--'Dashed,'-.'Dashed/dotted

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,10,8])
plt.plot(y,'o-.')
plt.show()
```

Color Reference('r'red,'g'Green,'b'blue,'c'cyan,'m'magenta)

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,10,8])
plt.plot(y,'o-m')
plt.show()
```

Marker Size (keyword argument markersize or the shorter version, ms)

```python
# Set the size of the markers to 10
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,10,8])
plt.plot(y,marker='o',ms=10)
plt.show()
```

```python
# keyword argument 'markeredgecolor' or the shorter 'mec'
# Set the EDGE color to magenta
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,10,8])
plt.plot(y,marker='o',ms=10,mec='m')
plt.show()
```

```python
# keyword argument 'markerfacecolor' or the shorter 'mfc'
# Set the FACE color to magenta
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,10,8])
plt.plot(y,marker='o',ms=10,mec='m',mfc='g')
plt.show()
```

```python
# Hexadecimal color values
# Mark each point with a beautiful green color
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,10,8])
plt.plot(y,marker='o',ms=15,mec='#4CAF50',mfc='#4CAF50')
plt.show()
```

```python
# The 140 supported color names
# Mark each point with the color named "hotpink"
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,10,8])
plt.plot(y,marker='o',ms=15,mec='hotpink',mfc='hotpink')
plt.show()
```
